```python
import pandas as pd
import random

random.seed(42)

characters = [
    "Messi","Ronaldo","Einstein","Newton","Tesla","IronMan","Batman",
    "Superman","Spiderman","Joker","Naruto","Goku","Luffy","Sasuke",
    "Doraemon","Mickey","Donald","ElonMusk","BillGates","SteveJobs",
    "Obama","Trump","Modi","Putin","TaylorSwift","Drake","Eminem",
    "Beyonce","Ariana","Rihanna","Shahrukh","Aamir","Salman",
    "Thor","Hulk","CaptainAmerica","Deadpool","Wolverine",
    "Flash","WonderWoman","BlackPanther","Aquaman","Vision",
    "Thanos","Ultron","Loki","Homelander","LightYagami"
]

attributes = [
    "real","fictional","male","female","movie","tv","anime","sports",
    "politics","music","scientist","alive","superpower","villain","animated"
]

data = []

for name in characters:
    row = {"character": name}
    for attr in attributes:
        row[attr] = random.randint(0, 1)
    data.append(row)

df = pd.DataFrame(data)
df.to_csv("akinator_data.csv", index=False)

print("CSV file created with shape:", df.shape)
```
